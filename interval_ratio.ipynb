Histogram for interval data visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


temperatures = np.random.normal(loc=20, scale=5, size=1000)
temperatures = np.clip(temperatures, a_min=-10, a_max=40)  # Limit temperatures between -10 and 40 degrees Celsius

df = pd.DataFrame({'Temperature': temperatures})

plt.figure(figsize=(10,6))
plt.hist(df['Temperature'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

Scatterplot with trendline

```python
import random
import matplotlib.pyplot as plt

# Generate sample temperature data (in Celsius)
temperatures = []
for _ in range(100):
    temp = random.uniform(-10, 40)  # Temperatures between -10 and 40 degrees Celsius
    temperatures.append(temp)

# Generate sample humidity data
humidities = []
for _ in range(100):
    humidity = random.uniform(20, 100)  # Humidity between 20% and 100%
    humidities.append(humidity)

# Calculate the means of temperatures and humidities
mean_temp = sum(temperatures) / len(temperatures)
mean_humidity = sum(humidities) / len(humidities)

# Calculate the numerator and denominator for the slope (m)
numerator = sum((temp - mean_temp) * (hum - mean_humidity) for temp, hum in zip(temperatures, humidities))
denominator = sum((temp - mean_temp) ** 2 for temp in temperatures)

# Avoid division by zero
if denominator != 0:
    slope = numerator / denominator
else:
    slope = 0

# Calculate the intercept (b)
intercept = mean_humidity - slope * mean_temp

# Create trendline values
trendline_temperatures = [min(temperatures), max(temperatures)]
trendline_humidities = [slope * temp + intercept for temp in trendline_temperatures]

# Scatter Plot with Trendline
plt.figure(figsize=(10, 6))
plt.scatter(temperatures, humidities, alpha=0.7, color='purple', label='Data Points')
plt.plot(trendline_temperatures, trendline_humidities, color='red', linewidth=2, label='Trendline')
plt.title('Temperature vs. Humidity with Trendline')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.legend()
plt.show()
```

Vertically aligned histogram to highlight patterns:

```python
import matplotlib.pyplot as plt
import random

# Generate datasets
male_heights = [random.gauss(70, 3) for _ in range(1000)]     # Mean=70 inches, SD=3 inches
female_heights = [random.gauss(65, 3) for _ in range(1000)]   # Mean=65 inches, SD=3 inches

# Create subplots aligned vertically
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot histograms
ax1.hist(male_heights, bins=20, color='blue', alpha=0.7)
ax2.hist(female_heights, bins=20, color='red', alpha=0.7)

# Set titles and labels
ax1.set_title('Male Heights')
ax2.set_title('Female Heights')
ax2.set_xlabel('Height (inches)')
ax1.set_ylabel('Frequency')
ax2.set_ylabel('Frequency')

# Ensure x-axis limits are the same for both histograms
ax1.set_xlim(50, 85)
ax2.set_xlim(50, 85)

# Add gridlines for better readability
ax1.grid(True)
ax2.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

Illusory relationship 

```python
import matplotlib.pyplot as plt
import random

# Generate interval data for x values
x_values = [i for i in range(100)]  # x-values: 0 to 99

# Generate random y values with noise, which could give the appearance of a relationship
y_values = [50 + random.uniform(-30, 30) + 0.2*random.choice([-1, 1])*x for x in x_values]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, color='purple')
plt.title('Scatter Plot: Illusory Relationship')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)

# Add trendline
# Calculate slope and intercept for trendline (using manual least squares calculation)
mean_x = sum(x_values) / len(x_values)
mean_y = sum(y_values) / len(y_values)
numerator = sum((x - mean_x) * (y - mean_y) for x, y in zip(x_values, y_values))
denominator = sum((x - mean_x) ** 2 for x in x_values)
slope = numerator / denominator if denominator != 0 else 0
intercept = mean_y - slope * mean_x

# Calculate trendline values
trendline_y = [slope * x + intercept for x in x_values]

# Plot the trendline
plt.plot(x_values, trendline_y, color='red', label=f'Trendline (slope: {slope:.2f})')
plt.scatter(x_values, y_values, color='red')
plt.legend()

# Show the plot
plt.show()
```

Spurious correlation

```python
import matplotlib.pyplot as plt
import random

# Simulate data for ice cream sales and drowning deaths (positive correlation due to a hidden factor - hot weather)
# Data for 12 summer months (e.g., over several years)

# Generate random data for ice cream sales (in thousands) and drowning deaths
ice_cream_sales = [random.uniform(50, 100) for _ in range(12)]  # Ice cream sales in thousands of units
drowning_deaths = [sales * random.uniform(0.4, 0.6) for sales in ice_cream_sales]  # Deaths correlated with sales

# Create scatter plot to show positive correlation
plt.figure(figsize=(8, 6))
plt.scatter(ice_cream_sales, drowning_deaths, color='orange', s=100)
plt.title('Positive Correlation: Ice Cream Sales vs. Drowning Deaths')
plt.xlabel('Ice Cream Sales (Thousands)')
plt.ylabel('Drowning Deaths')
plt.grid(True)

# Show plot
plt.show()

# Explanation message
print("Note: While there's a positive correlation between ice cream sales and drowning deaths, this doesn't imply that ice cream sales cause drowning deaths. \nIn reality, a hidden factor (hot weather) affects both variables.")
```
